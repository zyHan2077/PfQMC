```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

## $T = 0.2$, majorana resolved sign and RG ratio

```python
# pickle get data from file
data = None
with open('./data/data_honeycomb_L6-15_V1.0-2.0_len16k-new.pkl', 'rb') as f:
    data = pickle.load(f)

tmp = {}
for i in data:
    tmp[i[0]] = i[1]

data = tmp
```

```python
Vs = np.arange(1.0, 1.94, 0.05)
# Ls = np.arange(8, 19, 2)
Ls = [6, 9, 12, 15]

fig, axs = plt.subplots(2, 1, figsize=(9, 12))

for L in Ls:
    signs = []
    signsStd = []
    cdws = []
    cdwsStd = []
    cdwqs = []
    cdwqsStd = []
    rgRatio = []
    rgRatioStd = []

    for V in Vs:
        # print("L = {}, V = {}".format(L, V))
        # obs, params = getChainDataFromFile(L, V)
        # r = biningAndAnalysis(obs)
        r = data[(L, V)]
        signs.append(r["obsSign"][0])
        signsStd.append(r["obsSign"][1])
        cdws.append(r["obsCDW"][0])
        cdwsStd.append(r["obsCDW"][1])
        cdwqs.append(r["obsCDWq"][0])
        cdwqsStd.append(r["obsCDWq"][1])
        rgRatio.append(r["rgRatio"][0])
        rgRatioStd.append(r["rgRatio"][1])

    axs[0].errorbar(Vs, signs, yerr=signsStd, label=r"$L = {}$".format(L), marker='o', capsize=3)

    axs[1].errorbar(Vs, 1.0 - np.array(rgRatio), yerr=rgRatioStd, label=r"$L = {}$".format(L), marker='o', capsize=3)

axs[0].set_ylabel(r"resolved sign")
    # axs[0].set_ylim(0, 1.0)
# axs[0].set_yscale("log")
axs[0].legend()
axs[1].set_ylabel(r"RG ratio")
# axs[1].set_ylim(0, 1.0)
# axs[1].set_yscale("log")
axs[1].set_xlabel(r"$V$")
axs[1].legend()

fig.suptitle(r"$L\tau = 50, \Delta t = 0.1$")
plt.legend()
plt.show()
```
